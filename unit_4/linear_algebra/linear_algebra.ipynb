```python
def isDiagnol(matrix):
    if matrix.shape[0]!=matrix.shape[1]:
        return False 
    dim=matrix.shape[0]
    for i in range(dim):
        for j in range (i):
            if matrix[i,j]!=0 and !=j:
                return False 
    return True
```

```python
def isuppertriangle(matrix):
    if matrix.shape[0]!=matrix.shape[1]:
        return False 
    dim=matrix.shape[0]
    for i in range(dim):
        for j in range (i):
            if matrix.shape[i,j]!=0 :
                return False
    return True
```

```python
def islowertriangle(matrix):
    if matrix.shape[0]!=matrix.shape[1]:
        return False 
    dim=matrix.shape[0]
    for i in range(dim):
        for j in range (i+1,dim):
            if matrix.shape[i,j]!=0 :
                return False
    return True
```
